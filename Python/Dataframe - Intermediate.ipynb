```python
filter_data = data_baru.loc[(data_baru['age'] > 20) | (data_baru['length_of_service'] > 3) | (data_baru['region'] == "region_27"),["employee_id","department","region","age","length_of_service"]]
filter_data
```

**Creating New Variable**

```python
filter_data['test'] = filter_data['age']+filter_data['length_of_service']
filter_data.head()
```

```python
filter_data['gabungan'] = str(filter_data['department'])+str(filter_data['region'])
filter_data
```

**Grouping and Merging**

```python
filter_data.groupby('region').agg(rerata_usia= ("age","mean"),
                                  max_service=("length_of_service","max")).reset_index()
#rata2 usia dan lamanya length of service berdasarkan region
#.groupby() selalu bareng .agg()
#pake .reset_index() biar kolom region kedudukannya sama 
```
