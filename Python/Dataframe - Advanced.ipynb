```python
#join
data_1 = data.sample(n=5)
data_2 = data.sample(n=5)

display(data_1.head())
display(data_2.head())
```

```python
data_1.join(data_2, lsuffix='_first',rsuffix='_second')
```

```python
pd.concat([data_1,data_2],axis=1) #horizontal
```

```python
pd.concat([data_1,data_2],axis=0) #vertikal
```

```python
data_1.append(data_2)
```

```python
#pivot table
data.set_index(['sex','smoker'])
```

```python
#summary usia untuk berdasarkan sex dan smoker di region masing2
pd.pivot_table(data, values='age',index=['sex','smoker'], columns = 'region',
               aggfunc='mean') 
```

```python
pd.pivot_table(data, values='age',index=['sex','smoker'], columns = 'region',
               aggfunc=['mean','min','max']) #kalo aggfunc lebih dari 1
```

```python
data.groupby(['sex','smoker']).agg(average_age = ('age','mean'),
                                   min_age= ('age','min'))
```

```python
import time
start = time.time()

cost = []
for i in data['bmi']:
  if i < 30:
    cost.append('Low BMI')
  else:
    cost.append('High BMI')

end = time.time()
print(end - start)
```

```python
data['bmi_category'] = cost
data.head()
```

```python

data['bmi_cate_lambda'] = data['bmi'].apply(lambda x: 'Low BMI' if x <30 else 'High BMI')
data.head()
```
