##**Hypothesis Testing**

```python
data['CO2'].mean()
```

```python
#Ztest
avg_CO = data['CO2'].mean()
print(avg_CO)

working = data[data['event'] == 'Working']
sample_CO = working['CO2'].sample(n=35)


ztest, pvalue = ztest(x1=sample_CO, x2 =None, value=avg_CO, alternative ='larger' )

print('ztest: {} and pvalue {}'.format(ztest, pvalue))

CI = 0.95
alpha = 1-0.95

if pvalue < alpha:
  print('Reject Hypothesis')
  print("Pvalue{} < alpha{}".format(round(pvalue,2), round(alpha,2)))

else:
  print('Accept Hypothesis')
  print("Pvalue{} > alpha{}".format(round(pvalue,2), round(alpha,2)))
```

```python
# ttest
avg_O = data['Oxigen'].mean()
print(avg_O)

working = data[data['event'] == "Working"]
sample_O = working['Oxigen'].sample(n=25)


ttest, pvalue = stats.ttest_1samp(sample_O, avg_O, alternative='less')

print('ttest: {} and pvalue: {}'.format(ttest, pvalue))

CI = 0.95
alpha = 1-0.95

if pvalue < alpha:
  print("Reject Hypothesis")
  print("Pvalue {} < alpha {}".format(round(pvalue,2), round(alpha,2)))
  
else:
  print("Accept Hypothesis")
  print("Pvalue {} > alpha {}".format(round(pvalue,2), round(alpha,2)))
```

```python
chi = data[['event', 'CO',"CO2",'Hydrogen','Methane']]
chi = chi.set_index('event')
chi
```

```python
# Chi Square
from scipy.stats import chi2_contingency
stat, pvalue, dof, expected = chi2_contingency(chi)

print('Chi square: {} and pvalue: {}'.format(stat, pvalue))

CI = 0.95
alpha = 1-0.95

print()
if pvalue < alpha:
  print("Reject Hypothesis")
  print("Pvalue {:.5f} < alpha {}".format(round(pvalue,10), round(alpha,2)))
  
else:
  print("Accept Hypothesis")
  print("Pvalue {:.5f} > alpha {}".format(round(pvalue,10), round(alpha,2)))
```
